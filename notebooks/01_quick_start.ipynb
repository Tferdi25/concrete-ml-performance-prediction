# Quick Start: Concrete Performance Prediction

This notebook demonstrates the main workflow for training machine learning models to predict concrete mechanical, durability, and environmental performance indicators.


```python
import pandas as pd
from src.data import load_dataset, split_features_targets
from src.models import build_models
from src.config import TARGET_COLUMNS
from src.evaluate import regression_metrics
```

```python
data = load_dataset('../data/sample_concrete_dataset.csv')
data.head()
```

```python
x_train, x_test, y_train, y_test = split_features_targets(data)
models = build_models()
model = models['random_forest']
model.fit(x_train, y_train)
predictions = model.predict(x_test)
metrics = regression_metrics(y_test, predictions, TARGET_COLUMNS, 'random_forest')
metrics
```

## Interpretation

The table reports MAE, RMSE and R² for each target variable. In a real research project, the synthetic dataset should be replaced with experimental measurements from laboratory concrete mixtures.
